#### Introduction: Business Problem <a name="introduction"></a>

RRBP (Really Really Rich Business People Incorporated) approached the data science team with worries over the current economic situation in our country. Their fears have motivated them to plan an emigration of their major business resources and assets to London, where a large portion of business is conducted. They have requested that we provide them with possible locations for their offices to move. They would prefer to all live within the same borough, and thus it would need to have an active out-of-home and night life (restaurants, bars, outdoor activities... etc) as well as be relatively safe and have access to good schools within the area. The distance to city center is not a concern as most of their meetings are virtual, with only a few being in person. They have requested a submission of no less than five borough possibilities.

#### Data <a name="data"></a>

The problem requires that boroughs be assessed, grouped and submitted with motivating reasons for why those boroughs meet the definition of ideal.
Therefore the following requirements must be met:
* The data must assess the standard of living evidenced by the density of venues that match the description given - i.e. that there be a great deal of pubs, cafes, restaurants etc. 
* The data must include crime statistics.
* The data must include the number of good schools - which could be assessed by the number of independent schools in the area.

It was decided that a simple assessment of London Boroughs, the institutions present in those boroughs, and some basic statistics on crime and schools would be needed to find the best borough. 

The data was acquired through the following means:
* A JSON file of boroughs along with crime and school data was pulled from a github project by naomigg at https://github.com/naomiggg/the-best-borough-in-london/blob/master/boroughs.json.
* The Foursquare API was used to pull venue data for each borough.

For brevity's sake, the actions performed on the data are kept within the notebook where the code was run

#### Methodology <a name="methodology"></a>

The methodology undertaken mirrored the process taken in the lab provided by the IBM course on New York's Neighbourhoods. The differences were mainly in the final analysis of the clusters and the insights drawn from them.

Therefore, a map was built of the different boroughs to gain an idea of the relative location distribution. The average distance between boroughs was calculated in a rough manner, and so the radius selected for the venues to be included within was set at 1km, considering the boroughs are on average around 2km away from the nearest borough. 

The JSON file was cleaned into a dataframe, and venues were pulled for each borough. Counts were obtained for each category of venue, and the data was then scaled to allow for clustering later on through the ranking of relative importance through the count of venues. The clustering was then performed. A few sets of clusters were run, however the 2 largest clustering groups did not change very much and so 5 clusters were finalised. 

Following that, clusters were assessed, named, and assessed along with the crime and school data to provide a set of 5 possible boroughs for the RRBP. 

#### Analysis <a name="analysis"></a>

In the analysis of the results, we find that the 10 clusters are quite distinct.

1. Cluster 1 - City Tradepoint
2. Cluster 2 - City Dwelling
3. Cluster 3 - Unidentified - but there are very few pubs which is alarming.
4. Cluster 4 - Rural
5. Cluster 5 - Suburbs

Cluster 1 is identified by a single borough on the very Southern edge of London near an airport. It's likely that this particular borough is more of a travel entry point for those entering from the South. It is characterised by pubs, gas stations and supermarkets.

Cluster 2 is identified as the largest cluster, with most of London being fairly homogenous in its representation here. There is a large mix of pubs, cafes, restaurants, breweries and other interesting things to do such as golf courses. 

Cluster 3 is fairly unidentifiable as it is a single borough, however it has a pub nowhere within the top 10 locations, despite being near the waterside. This is starkly worrying as a cluster to look to for the purposes of this analysis.

Cluster 4 is a single rural borough on the outer edge of London, which is profused with farms, steakhouses and athletic locations. 

Cluster 5 is a cluster of suburbs. The venues in the vicinity are largely concerned with parks, supermarkets, restaurants and other amenable stores such as clothing stores. 

#### Results and Discussion <a name="results"></a>

In referring to the objectives laid out in the introduction, we can see that the most likely cluster to satisfy the RRBP is cluster 2 as it matches the proposition of an active night and out-of-the-house life. The next point is to find which suburbs in the cluster have the right balance of crime and schools to ensure fulfillment of the other requirements. Although crime typically is higher in poorer areas, in areas with high independent school numbers, the affluence means that sometimes the crime figure can go up due to there being easy pickings. So the correct balance needs to be struck.

Therefore the following boroughs are put up for the RRBP:
* Barnet - the most suburb like (meaning it qualifies less for the nightlife desired), however with its low crime rate and high number of good schools, it could fit the bill. 

* Richmond Upon Thames - has the lowest crime rate as well as a good number of pubs and french restaurants, as well as a decent amount of cultural sites to experience and possibly quite enjoyable expensive boat rides. 

* Kensington and Chelsea - has a high crime rate, however the activities outside the house are the best on the list in terms of balance between nightlife and being out during the day and matches the requirements best in that one regard.

* Brent - lower number of schools, however it does have the best athletic and outdoor activities such as soccer fields, tennis courts and parks as well as having sufficient pubs and bars for any business person.

* Hackney - has the strongest nightlife out of all suburbs and a high density of good schools, however it also has the highest crime rate of the selection. Note though, the crime is high in relation to the selected suburbs, there are suburbs with a far higher crime rate. 

These five suburbs were selected out of the number of suburbs in the second cluster as they had the best balance across the requirements set, and have enough variation that if the requirements did not fully capture the desires of the RRBP then they will certainly find what they want in there.


#### Conclusion <a name="conclusion"></a>

In this research task, the boroughs of London were investigated in their living standard as defined by requirements given by a business for the purpose of selecting the ideal suburb to emigrate to. Sticking to the brief given to the data science team, clusters of boroughs were made to determine which cluster of suburbs would align to the business's requirements. From there, the boroughs were further cut down to finally select 5 boroughs that best matched the business requirements. These requirements included the number of surrounding independent schools, a low crime rate and an active city life with things to do outside of the home. The final 5 boroughs selected were Barnet, Richmond Upon Thames, Kensington and Chelsea, Brent and Hackney. The final decision will be made by the stakeholders upon their own judgement and further research into the boroughs.
